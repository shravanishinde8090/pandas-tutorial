### impute missing values

#### somtimes if we delete entire column that will be not the appropriate approach.delete column can affect our model building we lose our main features.for imputing we ave many approach so here are some of the most popular techniques. 

### method 1 - impute fixed values like 0,'unknown' or'missing' etc we inpute unknown gender columns

df

```python
df['Gender'].fillna('Unknown', inplace=True)
df
```

### method 2- impute mean,median and mode

```python
# impute mean in amount_spend column
mean_amount_spent = df['Amount_spent'].mean()
df['Amount_spent'].fillna(mean_amount_spent, inplace=True)
df
mean_amount_spent
```

```python
#impute median in age column
median_Age = df['Age'].median()
df['Age'].fillna(median_Age, inplace=True)
df
median_Age
```

```python
#impute mode in employee_status column
mode_emp = df['Employees_status'].mode().iloc[0]
df['Employees_status'].fillna(mode_emp, inplace=True)
mode_emp
```

```python
mean_amount_spent
```

```python
mode_emp
```

### method 3- imputing forword fill or backfill by ffill ans bfill.in ffill missing value impute from the value of the above row and for bfill its taken from the below rows value.

```python
df['Referal'].fillna(method='ffill' , inplace=True)
df
```

```python
df.isna().sum().sum()
```

# 4.MEMORY MANAGEMENT

#### when we work on large dataset, there we get one big issue is a memory problem.we need too large resources for dealing with this.but there are some methods in pandas to deal with this, here are some methods or strategies to deal with the problem with help of pandas.

```python
df_memory = df.copy()
```

```python
memory_usage = df_memory.memory_usage(deep=True)
memory_usage_in_mbs = round(np.sum(memory_usage / 1024 ** 2), 3)
print(f" Total memory taking df_memory dataframe is : {memory_usage_in_mbs:.2f} MB ")
```

## change object to category datatype

#### our data frame is small in size. which is 1.15MB.now we will convert our object datatype to category

```python
# object datatype to category convert
```

```python
df_memory[df_memory.select_dtypes(['object']).columns] = df_memory.select_dtypes(['object']).apply(lambda x: x.astype('category'))
df_memory
```

```python
# convert object to category
df_memory.info(memory_usage="deep")
```

## change int64 or float64 to int32, 16, or 8

#### by default, pandas store numric values to int64 or float64.which takes more memory.if we have to store small numbers,then we can change to 64 to 32, 16 and so on. for example, our referal columns have only 0 and 1 values so for that we dont need to store at float64. so now we change it to float64

```python
# change referal column datatype
df_memory['Referal'] = df_memory['Referal'].astype('float32')
```

# 5.3 summarize or grouping data

## groupby

### grouping by one column: for example,if we want to find maximum values of age and Amount_spent by gender then we can use this:

```python
df[['Age','Amount_spent']].groupby(df['Gender']).max()
```

### to find mean,count,and max values of age and Amount_spent by gender then we can use age() function with groupby().

```python
# Gender by one columns
state_gender_res = df[['Age','Gender','Amount_spent']].groupby(['Gender']).agg(['count','mean','max'])
state_gender_res
```

### grouping by multiple columns: to find count,maximum and minimum values of amount_spent by State_names,gender,and payment_method then we can pass these columns name under groupby() function and add .agg() with count,mean,max argument

```python
# group by multiple columns
state_gender_res = df[['State_names','Gender','Payment_method','Amount_spent']].groupby([ 'State_names','Gender','Payment_method']).agg(['count','min','max'])
state_gender_res.head(12)
```
